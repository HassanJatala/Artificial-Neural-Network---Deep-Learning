# Loops

For Loop with range

```python
for i in range(10):
    print(i)
```

For loop with list

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"Fruit: {fruit}")
```

While loop

```python
counter = 0
while counter < 3:
    print("Counter: " ,counter)
    counter = counter + 1
```

Nested loop

```python
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

For-else

While-else

```python
num = 0
while num < 3:
    print(num)
    num += 1
else:
    print("loop done")
```

# Functions 

Simple function

```python
def greet(name):
    return f"Hello, {name}"

print(greet("Hassan"))
```

Function with default parameter

```python
def power(base, exponent=3):
    return base ** exponent

print("Power(3):",power(3))
print("Power(8):",power(8))
```

*args for variable number of arguments

```python
def sum_all(*numbers):
    return sum(numbers)

print(f"Sum: {sum_all(66, 67, 67, 67)}")
```

**kwargs for keyword arguments

```python
def person_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

person_info(name="Hassan", age=21, city="Faisalabad")
```

Recursive function

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(f"Factorial of 5: {factorial(5)}")
```

#  Lists

```python

list = [1, 2, 3, 4, 5]
print(f" list: {list}")

list.append(6)
print(f"1. After append(6): {list}")

list.insert(4, 67)
print(f"2. After insert(4, 67): {list}")

list.sort()
print(f"3. After sorting: {list}")

list.remove(67)
print(f"4. After remove(67): {list}")

popped = list.pop(3)
print(f"5. After pop(3): {list}, Popped: {popped}")

list.reverse()
print("6. List after reverse : " , list)

print("7. index no: ", list.index(3))

data = list.clear()
print("8. After clearing : " , data)
```

# Dictionaries

```python
print(" BASICS OF DICTIONARY")
student = {
    "name": "Hassan",
    "age": 21,
    "major": "Artificial Intelligence",
    "grades": {"stats": 90, "Analysis of Algo": 85, "Assembly": 88}
}
print(f"Student dictionary: {student}")
```

## LIST TO DICTIONARY

```python

keys = ['a', 'b', 'c']
values = [1, 2, 3]

dict_from_list = dict(zip(keys, values))
print(f"Dictionary from zip: {dict_from_list}")
```

## DICTIONARY TO LIST

```python
student = {
    "name": "Hassan",
    "age": 21,
    "major": "Artificial Intelligence",
    "grades": {"stats": 90, "Analysis of Algo": 85, "Assembly": 88}
}

keys_list = list(student.keys())
print(f"Keys as list: {keys_list}")

values_list = list(student.values())
print(f"Values as list: {values_list}")

items_list = list(student.items())
print(f"Items as list of tuples: {items_list}")
```

```python
print("TUPLES WITH DICTIONARIES")

coordinate_dict = {
    (0, 0): "Origin",
    (1, 2): "Point A",
    (3, 4): "Point B"
}
print(f"Dictionary with tuple keys: {coordinate_dict}")
print(f"Value at (1, 2): {coordinate_dict[(1, 2)]}")
```

## Dictionary

```python
print("DICTIONARY METHODS")

print(f"get('name'): {student.get('name')}")
print(f"get('nonexistent', 'default'): {student.get('nonexistent', 'default')}")


student.update({"year": "Junior", "gpa": 3.8})
print(f"After update: {student}")


popped_item = student.pop("age")
print(f"After pop('age'): {student}, Popped: {popped_item}")

last_item = student.popitem()
print(f"After popitem(): {student}, Last item: {last_item}")

student.setdefault("status", "Active")
print(f"After setdefault: {student}")
```

# NumPy

```python
import numpy as np
```

```python
print("NUMPY ARRAYS")

arr1 = np.array([1, 2, 3, 4, 5])
print(f"1D Array: {arr1}")
print(f"Type: {type(arr1)}")
print(f"Shape: {arr1.shape}")
print(f"Data type: {arr1.dtype}")

# 2D array
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\n2D Array:\n{arr2d}")
print(f"Shape: {arr2d.shape}")
```

```python

zeros_arr = np.zeros((3, 3))
print(f"\nZeros array:\n{zeros_arr}")

ones_arr = np.ones((2, 4))
print(f"\nOnes array:\n{ones_arr}")

identity_arr = np.eye(3)
print(f"\nIdentity matrix:\n{identity_arr}")

range_arr = np.arange(0, 10, 2)
print(f"\nArray from arange: {range_arr}")

linspace_arr = np.linspace(0, 1, 5)
print(f"\nLinspace array: {linspace_arr}")
```

## Vectorized Operations

```python
print("VECTORIZED OPERATIONS")

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print(f"Array a: {a}")
print(f"Array b: {b}")


print(f"\nElement-wise addition: {a + b}")
print(f"Element-wise subtraction: {a - b}")
print(f"Element-wise multiplication: {a * b}")
print(f"Element-wise division: {a / b}")
print(f"Element-wise exponent: {a ** 2}")
print(f"Element-wise square root: {np.sqrt(a)}")


print(f"\nElement-wise comparison (a > 2): {a > 2}")
print(f"Element-wise comparison (a == b): {a == b}")

print(f"\nSum of a: {np.sum(a)}")
print(f"Mean of a: {np.mean(a)}")
print(f"Standard deviation of a: {np.std(a)}")
print(f"Minimum of a: {np.min(a)}")
print(f"Maximum of a: {np.max(a)}")
print(f"Index of max value: {np.argmax(a)}")
```

# Matrix Operations and Broadcasting

```python
print("MATRIX MULTIPLICATION")

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)
print(result)
```

```python

mat=np.array([[1,2],[3,4]])
vec=np.array([10,20])
print("Matrix Multiplication with vectors : " , np.matmul(mat,vec))


print(f"\nMatrix A:")
print(f"Shape: {A.shape}")
print(f"Determinant: {np.linalg.det(A)}")
print(f"Transpose:\n{A.T}")
print(f"Inverse:\n{np.linalg.inv(A)}")
```
